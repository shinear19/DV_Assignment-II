```python
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt

# Fetch Apple stock data
apple_stock = yf.download("AAPL", start="2022-01-01", end="2023-01-01")

# Extract data for plotting: Daily returns and trading volume
apple_stock['Daily Return'] = apple_stock['Adj Close'].pct_change()
data = apple_stock[['Daily Return', 'Volume']].dropna()

# Ensure the data is 1-dimensional (flattened)
daily_return = data['Daily Return'].values.flatten()
volume = data['Volume'].values.flatten()

# Plot bivariate distribution: Daily Return vs. Volume
plt.figure(figsize=(10, 6))
sns.kdeplot(
    x=daily_return,
    y=volume,
    fill=True,
    cmap='Blues',
    levels=30
)

plt.title('Bivariate Distribution: Daily Return vs. Volume')
plt.xlabel('Daily Return')
plt.ylabel('Volume')
plt.grid()
plt.show()
```
